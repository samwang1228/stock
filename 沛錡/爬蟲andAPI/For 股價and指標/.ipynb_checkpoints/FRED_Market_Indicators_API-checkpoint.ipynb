```python
# API key https://fred.stlouisfed.org/docs/api/api_key.html
# API Github https://github.com/mortada/fredapi
# 小示範 https://havocfuture.tw/blog/python-economy-fred
```

```python
from fredapi import Fred
# 設定 API Key 需註冊申請自己的  
api_key = ''
fred = Fred(api_key=api_key)
data = fred.get_series('DGS10')  # 美國十年期公債殖利率

print(data)
```

```python
#test
print(data['1962-01-02'])
print(type(data))
```

```python
#轉換
#to dict
dict_data=data.to_dict()
#print(dict_data)

#to csv
compression_opts = dict(method='zip',
                        archive_name='FRED_API_demo_result.csv')  
data.to_csv('FRED_API_demo_result.zip', index=True,
          compression=compression_opts) 

csv_data=data.to_csv()
print(csv_data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

data.plot()
plt.grid()
plt.show()
```
