```python
import requests
import pandas as pd
url = "https://api.finmindtrade.com/api/v4/data"
parameter = {
    "dataset": "TaiwanStockNews",
    "data_id": "2498",  # 宏達電
    "start_date": "2022-04-15",
    "end_date": "2022-07-05",
    # 登入，獲取金鑰 https://finmind.github.io/quickstart/
    "token": "",
}
data = requests.get(url, params=parameter)
data = data.json()
data = pd.DataFrame(data['data'])
```

```python
print(data["title"])
```

```python
# 擷取各家新聞內容需要個別分析與爬取
# 中時新聞網 Chinatimes.com, ETtoday財經雲, Apple Daily TW 三個擷取資料方式已完成
```

```python
import requests
```

```python
# 中時新聞網 Chinatimes.com
```

```python
my_headers = {'cookie': 'over18=1;'}
url = data['link'][0]
response = requests.get(url, headers=my_headers)
content = response.text
print(data['source'][0])
print(content)
```

```python
print(url)
```

```python
import bs4
```

```python
root = bs4.BeautifulSoup(content, "html.parser")
paragraphs = root.find_all('p')
for paragraph in paragraphs:
    if paragraph.string != None:
        print(paragraph.string)
```

```python
# ETtoday財經雲
```

```python
my_headers = {'cookie': 'over18=1;'}
url = data['link'][2]
response = requests.get(url, headers=my_headers)
content = response.text
print(data['source'][2])
print(content)
```

```python
print(url)
```

```python
root = bs4.BeautifulSoup(content, "html.parser")
article = root.find_all('div', class_='subject_article')
print(article)
```

```python
paragraphs = article[0].find_all('p')
for paragraph in paragraphs:
    if paragraph.string != None:
        print(paragraph.string)
```
