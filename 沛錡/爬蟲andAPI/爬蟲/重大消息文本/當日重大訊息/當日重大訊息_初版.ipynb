```python
import requests
import pandas as pd
```

```python
import bs4
```

```python
url='https://mops.twse.com.tw/mops/web/t05st02'
```

```python
year="110"
month="2"
day="1"
```

```python
payload = {
        "encodeURIComponent": "1",
        "step": "1",
        "step00": "0",
        "firstin": "1",
        "off": "1",
        "TYPEK": "all",
        "year": year,
        "month": month,
        "day": day,
   }
```

```python
response = requests.post(url, data = payload)
raw_data=response.text
print(raw_data)
```

```python
soup = bs4.BeautifulSoup(raw_data, 'html.parser')
```

```python
rowDatasEven = soup.find_all('tr',class_='even')
#print(rowDatasEven)
```

```python
rowDatasOdd = soup.find_all('tr',class_='odd')
#print(rowDatasOdd)
```

```python
rowDatas=rowDatasEven+rowDatasOdd
#print(rowDatas[7])
```

```python
data = {'date': [], 'time': [], "title": [], 'content': [], 'co_id': [], 'date_of_event': []}
#col_name = {'date', 'time', "title", 'content', 'co_id', 'date_of_event'}
```

```python
#擷取資料
```

```python
columns=rowDatas[0].find_all('td')

colDatas=[]
for col in columns[5].find_all('input'):
    colDatas.append(col.get('value'))
#print(colDatas)
```

```python
#整理資料
```

```python
#date 發布日期
data['date'].append(colDatas[2])
#print(data['date'])

#time 發布時間
time=colDatas[3]
if len(time)==5:
    time='0'+time
data['time'].append(time)
#print(data['time'])

#title 標題
data['title'].append(colDatas[4])
#print(data['title'])

#content 內文
data['content'].append(colDatas[8])
#print(data['content'])


#co_id 股票代號
data['co_id'].append(colDatas[1])
#print(data['co_id'])

#date_of_event 事實發布日期
data['date_of_event'].append(colDatas[7])
#print(data['date_of_event'])
```

```python
d=pd.DataFrame(data)
```
