```python
import pandas as pd
from gensim.models.word2vec import Word2Vec
```

```python
# 讀取文本斷詞結果
pickle_df = pd.read_pickle('w2v_data/w2v_full_input.pkl')
```

```python
# 修正一下 發現資料有沒清乾淨的字元
new_list = []
for sentence in pickle_df['text']:
    new_sentence = []
    for word in sentence:
        word = word.replace('\n', '')
        word = word.replace(' ', '')
        new_sentence.append(word)
    new_list.append(new_sentence)
pickle_df['text'] = new_list
```

```python
corpus = pickle_df
model = Word2Vec(corpus.text, min_count=1, vector_size=80,
                 window=5, epochs=5)  # 向量長度設為80
```

```python
def most_similar(w2v_model, words, topn=10):
    similar_df = pd.DataFrame()
    for word in words:
        try:
            similar_words = pd.DataFrame(w2v_model.wv.most_similar(
                word, topn=topn), columns=[word, 'cos'])
            similar_df = pd.concat([similar_df, similar_words], axis=1)
        except:
            print(word, "not found in Word2Vec model!")
    return similar_df
```

```python
# 驗證一下結果好壞
most_similar(model, ['兆豐', '收支', '仟', '說明會', '盈餘', '董座', '財報'])
```

```python
model.save("w2v_full.model")
```

```python
# model格式轉換
"""程式碼來源:
https://sfhsu29.medium.com/nlp-%E5%B0%88%E6%AC%84-1-2-%E5%A6%82%E4%BD%95%E8%A8%93%E7%B7%B4%E8%87%AA%E5%B7%B1%E7%9A%84-word2vec-5a0754c5cb09
"""

# 由於這個 word2vec model 是專門給 gensim 用的，如果要給 keras使用，要先將model儲存成不同的格式
import numpy as np
from gensim.models.word2vec import Word2Vec
w2v_model = Word2Vec.load("w2v_data/w2v_full.model")
w2v_model.wv.save_word2vec_format('keras_word2vec.txt', binary=False)

# load the whole embedding into memory
embeddings_index = dict()
with open('keras_word2vec.txt', encoding='UTF-8') as f:
    for line in f:
        values = line.split()
        # 只接受長度為 81 的向量 (word + 80d embedding)
        if len(values) != 81:
            print(values)
            continue
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
```
