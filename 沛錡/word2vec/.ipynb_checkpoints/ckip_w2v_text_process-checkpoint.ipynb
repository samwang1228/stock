```python
import csv
import numpy as np
import pandas as pd
from gensim.models.word2vec import Word2Vec
```

```python
#斷詞的部分
```

```python
file = open('info_data/2317Labeldata.csv', encoding="utf-8")
reader = csv.reader(file)
data_list = list(reader)
file.close()
```

```python
#去標點符號、數字
from zhon.hanzi import punctuation
import string
punctuation_str=punctuation+string.punctuation
number='0123456789'
for text in data_list:
    #去標點符號
    for i in punctuation_str:
        text[1]=text[1].replace(i, ' ')
    #去數字
    for i in number:
        text[1]=text[1].replace(i, '')
    text[1]=text[1].replace('\t', '')
```

```python
#轉成dict
label_text_dict={'Info':[],'Label':[],'Words':[],'Vectors':[]}
for row in data_list[1:]:
    label_text_dict['Info'].append(row[1])
    label_text_dict['Label'].append(row[2])
```

```python
def extract_ckip_word(ws_list,pos_list):
    pos_remove_list=['Ne','WHITESPACE','Ca','DE','P','T','I']
    result=[]
    for i in range(len(ws_list)):
        temp_sentence=[]
        for j in range(len(ws_list[i])):
            for remove_pos in pos_remove_list:
                if remove_pos in pos_list[i][j]:
                    #print(remove_pos)
                    is_remove_pos=True
                    break
                else:
                    is_remove_pos=False
            if is_remove_pos==False:
                temp_sentence.append(ws_list[i][j])
        #if temp_sentence!=[]:
        result.append(temp_sentence)
    return result
def save_data(data_dict,name):
    data_df = pd.DataFrame(data_dict)
    data_df.to_pickle(name+'.pkl')
```

```python
#斷詞
from ckiptagger import WS, POS, NER
ws_list=[]
pos_list=[]
ws = WS("./data")
pos = POS("./data")
count=0
sentences=label_text_dict['Info']
for text in sentences: 
    ws_results = ws([text])
    pos_results = pos(ws_results)
    ws_list=ws_list+ws_results
    pos_list=pos_list+pos_results
    count+=1
    if count%200==0:
        print(count) 
```

```python
label_text_dict['Words']=extract_ckip_word(ws_list,pos_list)
```

```python
#修正一下 發現資料有沒清乾淨的字元
new_list=[]
for sentence in label_text_dict['Words']:
    new_sentence=[]
    for word in sentence:
        word=word.replace('\n', '')
        word=word.replace(' ', '')
        new_sentence.append(word)
    new_list.append(new_sentence)
label_text_dict['Words']=new_list
```

```python
for sentence in label_text_dict['Words']:
    for word in sentence:
        if ' ' in word:
            print(word)
```

```python
#先儲存一下
save_data(label_text_dict,'2317_ckip_result')
```

```python
#Word2vec轉換部分 有斷詞的檔案可以直接跑這裡
```

```python
pickle_data=pd.read_pickle('2317_ckip_result.pkl')
label_text_dict={'Info':[],'Label':[],'Words':[],'Vectors':[]}

#直接使用pickle_data會跳奇怪的error 所以重新轉換一下
for i in range(len(pickle_data['Info'])):
    label_text_dict['Info'].append(pickle_data['Info'][i])
    label_text_dict['Label'].append(pickle_data['Label'][i])
    label_text_dict['Words'].append(pickle_data['Words'][i])
label_text_dict['Vectors']=[]
```

```python
#type(pickle_data['Words'][i])
```

```python
model= Word2Vec.load("w2v_data/w2v_750000.model")
```

```python
sentence_data=[]
for line in range(len(label_text_dict['Words'])):
    sentence_data.append(label_text_dict['Words'][line])
```

```python
#每個詞轉換成vector
for words in sentence_data:
    sentence_vectors=[]
    for word in words:
        try:
            sentence_vectors.append(model.wv[word])
            #print(model.wv[word])
        except:
            sentence_vectors.append(np.array([0]))
    label_text_dict['Vectors'].append(sentence_vectors)
#確認資料對應都正確
print(len(label_text_dict['Vectors'])) 
print(len(label_text_dict['Words']))
print(len(label_text_dict['Label']))
```

```python
#儲存結果
save_data(label_text_dict,'2317_vector_data_750000')
```
