```python
import pandas as pd
from gensim.models.word2vec import Word2Vec
```

```python
pickle_df = pd.read_pickle('w2v_data/w2v_input_750000.pkl')
```

```python
#修正一下 發現資料有沒清乾淨的字元
new_list=[]
for sentence in pickle_df['text']:
    new_sentence=[]
    for word in sentence:
        word=word.replace('\n', '')
        word=word.replace(' ', '')
        new_sentence.append(word)
    new_list.append(new_sentence)
pickle_df['text']=new_list
```

```python
corpus=pickle_df
model = Word2Vec(corpus.text,min_count=1,vector_size=80,window=5,epochs=5)
```

```python
def most_similar(w2v_model, words, topn=10):
    similar_df = pd.DataFrame()
    for word in words:
        try:
            similar_words = pd.DataFrame(w2v_model.wv.most_similar(word, topn=topn), columns=[word, 'cos'])
            similar_df = pd.concat([similar_df, similar_words], axis=1)
        except:
            print(word, "not found in Word2Vec model!")
    return similar_df
```

```python
#test
most_similar(model,['元大','收支','說明會','盈餘','董事會','財報','除息','仲裁'])
```

```python
model.save("w2v_750000.model")
```

```python
#load model method
model2= Word2Vec.load("w2v_data/w2v_750000.model")

#test
most_similar(model2,['兆豐','收支','仟','說明會','盈餘','董座','財報'])
```
