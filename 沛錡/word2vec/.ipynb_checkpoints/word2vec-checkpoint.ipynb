```python
import pandas as pd
from gensim.models.word2vec import Word2Vec
```

```python
pickle_df = pd.read_pickle('word2vec_input.pkl')
```

```python
corpus=pickle_df
model = Word2Vec(corpus.text,min_count=1,vector_size=50,window=10,epochs=5)
```

```python
def most_similar(w2v_model, words, topn=15):
    similar_df = pd.DataFrame()
    for word in words:
        try:
            similar_words = pd.DataFrame(w2v_model.wv.most_similar(word, topn=topn), columns=[word, 'cos'])
            similar_df = pd.concat([similar_df, similar_words], axis=1)
        except:
            print(word, "not found in Word2Vec model!")
    return similar_df
```

```python
#test
most_similar(model,['兆豐','收支','仟','說明會','盈餘','董座','財報'])
```

```python
model.save("w2v.model")
```

```python
#load model method
model2= Word2Vec.load('w2v.model')

#test
most_similar(model2,['兆豐','收支','仟','說明會','盈餘','董座','財報'])
```
