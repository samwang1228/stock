get_income_statement() -> 爬取單筆資料

saveDataToExcel(stockID,start,end)-

```python
def get_income_statement(sym, year, season): 
    url = 'https://mops.twse.com.tw/mops/web/ajax_t164sb04'
    payload = {
               'encodeURIComponent': '1',
               'step': '1',
               'firstin': '1',
               'off': '1',
               'queryName': 'co_id',
               'inpuType': 'co_id',
               'TYPEK': 'all',
               'isnew': 'false',
               'co_id': str(sym),
               'year': str(year),
               'season': str(season),
           }
    res = requests.post(url , data = payload)
    df = pd.read_html(res.text)[1]
    return df
```

```python
import time
# df=get_income_statement(2330,110,2)
def saveDataToExcel(stockID,start,end):
    interval=end-start+1 
    mySheet=[]
    df=[]
    counter=0
    with pd.ExcelWriter(f'{stockID}.xlsx', engine='xlsxwriter') as writer:
        for i in range(interval):
            for j in range(4):
                try:
                    df.append(get_income_statement(stockID, end-i, 4-j))
                    mySheet.append(f'民國第{end-i}年第{4-j}季')
                    df[counter].to_excel(writer, sheet_name=mySheet[counter])
                    counter=counter+1
                    time.sleep(20)
                except:
                    print(f'民國第{end-i}年第{4-j}季 no data')
    counter=0
    with pd.ExcelWriter(f'{stockID}.xlsx', engine='xlsxwriter') as writer:
        for subSheet in mySheet: 
                df[counter].to_excel(writer, sheet_name=subSheet, index=True, na_rep='NaN')
                # Auto-adjust columns' width
                for column in df[counter]:
                    column_width = 40
                    col_idx = df[counter].columns.get_loc(column)
                    writer.sheets[subSheet].set_column(col_idx, col_idx, column_width)
                counter=counter+1
    #     writer.save()
    print("File loading is finsh!")
saveDataToExcel(2330,111,110)
```
