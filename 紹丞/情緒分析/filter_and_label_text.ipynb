# merge all code

```python
substr=''
ans=[]
label=[]
sublabel=0
for i in range(len(df)):
    temp=df['content'][i]
    substr=temp
    counter=2
    pre=-1
#     string to time format
    subtime=str(df['date'][i])
    subtime=subtime[0:4]+'-'+subtime[4:6]+'-'+subtime[6:8]
    subtime=datetime.datetime.strptime(subtime, "%Y-%m-%d")
#     ----------
#   To compare price after announcing info time
    condi=stock_price[(stock_price.Date==subtime)].index.tolist()
#     print(condi[0])
    try:
        sublabel=(stock_price.iloc[condi[0]+3]['Close']-stock_price.iloc[condi[0]]['Close'])/stock_price.iloc[condi[0]]['Close']
        sublabel*=10
    except:
        sublabel=0
#   threshold 
    if(sublabel>1):
        sublabel=1
    elif(sublabel<-1):
        sublabel=-1
    elif(sublabel<0.3 and sublabel>-0.2):
        sublabel=0
#     timeList.append(subtime)
# ---------------------
    for j in range(len(substr)-1):
#             print(j)
        if(substr[j+1]==str(counter) and substr[j]=='\n'):
            counter+=1
            ans.append(substr[pre+1:j])
            label.append(sublabel)
            pre=j
#         大於10的情況
        elif(counter>=10 and substr[j]==str(counter//10) and substr[j+1]==str(counter%10) and substr[j-1]=='\n'):
            counter+=1
            if counter==11:
                ans.append(substr[pre+1:j-1])
                label.append(sublabel)
            else:
                ans.append(substr[pre:j-1])
                label.append(sublabel)
            pre=j
        # final line
        elif(j==len(substr)-2):
            if(counter<10):
                ans.append(substr[pre+1:j+2])
                label.append(sublabel)
            else:
                ans.append(substr[pre:j+2])
                label.append(sublabel)
        
#   
ans,label
```

```python
df_ans=pd.DataFrame(ans,columns=['Info'])
df_label=pd.DataFrame(label,columns=['Label'])
df_ans=df_ans.merge(df_label, how='inner', left_index=True, right_index=True)
df_ans
```
